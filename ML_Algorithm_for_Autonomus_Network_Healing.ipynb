```python
import os, time, threading, csv, random
from flask import Flask, request, Response, jsonify
from werkzeug.serving import make_server
import requests
from prometheus_client import CollectorRegistry, Counter, generate_latest
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import numpy as np
import socket

OUTPUT_DIR = "output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def get_free_port():
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.bind(('', 0))
    port = s.getsockname()[1]
    s.close()
    return port

LB_PORT = get_free_port()
NUM_BACKENDS = 3
BACKEND_PORTS = [get_free_port() for _ in range(NUM_BACKENDS)]

print("Selected ports:")
print("LB_PORT =", LB_PORT)
for i,p in enumerate(BACKEND_PORTS, start=1):
    print(f"web{i} -> {p}")

# parameters
BACKEND_BASE_PORT = BACKEND_PORTS[0]
ATTACKER_RATE_ATTACK = 300
ATTACK_DURATION = 18
MONITOR_INTERVAL = 2
BASELINE_DURATION = 20


def run_backend(port, instance_id, stop_event, error_rate_override=0.01):
    app = Flask(f"web-{instance_id}")
    registry = CollectorRegistry()
    REQ = Counter("http_requests_total", "Total HTTP requests", ["instance"], registry=registry)
    ERR = Counter("http_requests_errors_total", "HTTP error responses", ["instance"], registry=registry)

    @app.route("/")
    def index():
        delay = random.uniform(0.005, 0.02)
        time.sleep(delay)
        if random.random() < error_rate_override:
            ERR.labels(instance=instance_id).inc()
            REQ.labels(instance=instance_id).inc()
            return "error", 500
        REQ.labels(instance=instance_id).inc()
        return jsonify({"instance": instance_id, "ok": True})

    @app.route("/metrics")
    def metrics():
        return Response(generate_latest(registry), mimetype="text/plain")

    max_attempts = 3
    attempt = 0
    while attempt < max_attempts and not stop_event.is_set():
        try:
            server = make_server("0.0.0.0", port, app, threaded=True)
            server.timeout = 1
            print(f"[backend {instance_id}] bound on port {port}")
            break
        except OSError as e:
            attempt += 1
            print(f"[backend {instance_id}] bind failed on port {port}: {e}; picking new free port (attempt {attempt})")
            port = get_free_port()
            time.sleep(0.1)
    else:
        print(f"[backend {instance_id}] failed to bind after {max_attempts} attempts - aborting backend start")
        return

    while not stop_event.is_set():
        server.handle_request()
    server.server_close()
    print(f"[backend {instance_id}] stopped")

# ---------- Simple LB ----------
class SimpleLB(threading.Thread):
    def __init__(self, port, backends):
        super().__init__(daemon=True)
        self.port = port
        self.backends = list(backends)  # list of (host,port,instance)
        self.lock = threading.Lock()
        self.blacklist = set()
        self.app = Flask("lb")
        self.server = None
        self.idx = 0

        @self.app.route("/")
        def proxy_root():
            client_ip = request.remote_addr or "attacker"
            if "src" in request.args:
                client_ip = request.args.get("src")
            if client_ip in self.blacklist:
                return "blocked", 403
            backend = self._next_backend()
            url = f"http://{backend[0]}:{backend[1]}/"
            try:
                r = requests.get(url, timeout=2)
                return (r.content, r.status_code, r.headers.items())
            except Exception as e:
                return f"backend error: {e}", 502

        @self.app.route("/list")
        def list_backends():
            with self.lock:
                return jsonify([{"host": b[0], "port": b[1], "instance": b[2]} for b in self.backends])

    def _next_backend(self):
        with self.lock:
            if not self.backends:
                raise RuntimeError("no backends")
            b = self.backends[self.idx % len(self.backends)]
            self.idx += 1
            return b

    def block_ip(self, ip):
        with self.lock:
            self.blacklist.add(ip)
        print(f"[LB] blocked {ip}")

    def add_backend(self, host, port, instance):
        with self.lock:
            self.backends.append((host, port, instance))
        print(f"[LB] added backend {instance}:{port}")

    def remove_backend(self, instance):
        with self.lock:
            self.backends = [b for b in self.backends if b[2] != instance]
        print(f"[LB] removed backend {instance}")

    def run(self):
        # bind LB with retry if necessary
        attempt = 0
        max_attempts = 5
        port = self.port
        while attempt < max_attempts:
            try:
                self.server = make_server("0.0.0.0", port, self.app, threaded=True)
                self.server.timeout = 1
                print(f"[LB] bound on port {port}")
                break
            except OSError as e:
                print(f"[LB] bind failed on {port}: {e} -> trying new free port")
                port = get_free_port()
                attempt += 1
                time.sleep(0.1)
        if not self.server:
            raise RuntimeError("LB failed to bind to any port")
        # update port so external code can see actual LB port if it changed
        self.port = port
        while True:
            self.server.handle_request()

# ---------- MLMonitor (IsolationForest) ----------
class MLMonitor(threading.Thread):
    def __init__(self, lb, monitor_interval=MONITOR_INTERVAL, baseline_secs=BASELINE_DURATION):
        super().__init__(daemon=True)
        self.lb = lb
        self.interval = monitor_interval
        self.baseline_secs = baseline_secs
        self.metrics_log = []
        self.actions_log = []
        self._stop = threading.Event()
        self.model = None
        self.model_trained = False
        self.contamination = 0.05

    def _train_initial(self):
        arr = np.array([[r, e] for (_, r, e) in self.metrics_log])
        if len(arr) < 5:
            return False
        self.model = IsolationForest(contamination=self.contamination, random_state=42)
        self.model.fit(arr)
        self.model_trained = True
        print("[ML] trained IsolationForest on baseline samples:", arr.shape[0])
        return True

    def remediate_ddos(self, attacker_ip):
        ts = time.time()
        if attacker_ip:
            self.lb.block_ip(attacker_ip)
            self.actions_log.append((ts, f"blocked {attacker_ip}"))
        with self.lb.lock:
            if self.lb.backends:
                inst = self.lb.backends[0][2]
                self.lb.remove_backend(inst)
                self.actions_log.append((ts, f"removed {inst}"))
                new_inst = f"{inst}-r"
                new_port = get_free_port()
                ev = threading.Event()
                t = threading.Thread(target=run_backend, args=(new_port, new_inst, ev), daemon=True)
                t.start()
                time.sleep(0.2)
                self.lb.add_backend("127.0.0.1", new_port, new_inst)
                self.actions_log.append((ts, f"started {new_inst}:{new_port}"))
        extra_inst = f"web-ext-{int(time.time())%10000}"
        extra_port = get_free_port()
        ev2 = threading.Event()
        t2 = threading.Thread(target=run_backend, args=(extra_port, extra_inst, ev2), daemon=True)
        t2.start()
        time.sleep(0.2)
        self.lb.add_backend("127.0.0.1", extra_port, extra_inst)
        self.actions_log.append((time.time(), f"scaled {extra_inst}:{extra_port}"))
        print("[REM] remediation actions applied")

    def adaptive_feedback(self, success):
        if success:
            self.contamination = max(0.01, self.contamination * 0.95)
        else:
            self.contamination = min(0.2, self.contamination * 1.1)
        if len(self.metrics_log) >= 10:
            arr = np.array([[r, e] for (_, r, e) in self.metrics_log])
            self.model = IsolationForest(contamination=self.contamination, random_state=42)
            self.model.fit(arr)
            print(f"[ML] adaptive retrain done. contamination={self.contamination:.3f}")

    def stop(self):
        self._stop.set()

    def run(self):
        print("[MON] collecting baseline metrics...")
        baseline_end = time.time() + self.baseline_secs
        while time.time() < baseline_end and not self._stop.is_set():
            time.sleep(self.interval)
        print("[MON] baseline collection finished; waiting for training data...")
        while len(self.metrics_log) < max(5, int(self.baseline_secs / self.interval)) and not self._stop.is_set():
            time.sleep(0.2)
        self._train_initial()
        print("[MON] entering monitoring loop")
        window_len = max(1, int(self.baseline_secs / self.interval))
        while not self._stop.is_set():
            if len(self.metrics_log) < 1:
                time.sleep(self.interval); continue
            reqs = sum([r for (_, r, _) in self.metrics_log[-window_len:]])
            errs = sum([e for (_, _, e) in self.metrics_log[-window_len:]])
            ts = time.time()
            with open(os.path.join(OUTPUT_DIR, "ml_metrics_log.csv"), "a") as f:
                f.write(f"{ts},{reqs},{errs}\n")
            score, is_anom = (0.0, False)
            if self.model_trained:
                x = np.array([[reqs, errs]])
                score = float(self.model.decision_function(x)[0])
                is_anom = self.model.predict(x)[0] == -1
            self.actions_log.append((ts, "eval", reqs, errs, score, is_anom))
            if is_anom:
                print(f"[ML] anomaly detected (score={score:.4f}). remediate.")
                self.remediate_ddos("10.0.0.5")
                time.sleep(self.interval * 3)
                try:
                    recent = list(pd.read_csv(os.path.join(OUTPUT_DIR,"metrics_log.csv"), names=["ts","reqs","errs"]).tail(5)["reqs"])
                    recovered = np.median(recent) < max(40, np.mean(recent)*0.7)
                except Exception:
                    recovered = True
                self.adaptive_feedback(success=recovered)
                self.actions_log.append((time.time(), "post_remediate_recovered", bool(recovered)))
            time.sleep(self.interval)

# ---------- Collector ----------
def collector_thread(lb, monitor, stop_event):
    prev = {}
    while not stop_event.is_set():
        total_reqs = 0
        total_errs = 0
        for b in list(lb.backends):
            host, port, inst = b
            try:
                r = requests.get(f"http://{host}:{port}/metrics", timeout=1)
                txt = r.text
                req_count = 0
                err_count = 0
                for line in txt.splitlines():
                    if line.startswith('http_requests_total'):
                        try:
                            req_count = int(float(line.split()[-1]))
                        except:
                            pass
                    if line.startswith('http_requests_errors_total'):
                        try:
                            err_count = int(float(line.split()[-1]))
                        except:
                            pass
                prev_r = prev.get(inst, 0)
                prev_e = prev.get(inst+"_e", 0)
                delta_r = max(0, req_count - prev_r)
                delta_e = max(0, err_count - prev_e)
                prev[inst] = req_count
                prev[inst+"_e"] = err_count
                total_reqs += delta_r
                total_errs += delta_e
            except Exception:
                pass
        ts = time.time()
        with open(os.path.join(OUTPUT_DIR,"metrics_log.csv"), "a") as f:
            f.write(f"{ts},{total_reqs},{total_errs}\n")
        monitor.metrics_log.append((ts, total_reqs, total_errs))
        if len(monitor.metrics_log) > 500:
            monitor.metrics_log.pop(0)
        time.sleep(MONITOR_INTERVAL)

# ---------- Attacker ----------
def attacker_sim(lb_addr, duration, rate, src_ip="10.0.0.5"):
    end = time.time() + duration
    session = requests.Session()
    interval = 1.0 / max(1, rate)
    sent = 0
    while time.time() < end:
        try:
            session.get(f"http://{lb_addr}/?src={src_ip}", timeout=3)
        except:
            pass
        sent += 1
        time.sleep(interval)
    print(f"[attacker] finished (~{sent} requests)")

# ---------- Demo orchestration ----------
def run_demo():
    backend_threads = []
    backend_events = {}
    backends = []
    for i, port in enumerate(BACKEND_PORTS):
        inst = f"web{i+1}"
        ev = threading.Event()
        t = threading.Thread(target=run_backend, args=(port, inst, ev), daemon=True)
        t.start()
        backend_threads.append(t)
        backend_events[inst] = ev
        backends.append(("127.0.0.1", port, inst))
        time.sleep(0.05)

    lb = SimpleLB(LB_PORT, backends)
    lb.start()
    # give LB a moment to bind (it may pick a different port if initial LB_PORT was busy)
    time.sleep(0.5)
    print("LB running on port:", lb.port, "   (LB_PORT variable used initially was", LB_PORT, ")")

    monitor = MLMonitor(lb)
    monitor.start()

    stop_col = threading.Event()
    col = threading.Thread(target=collector_thread, args=(lb, monitor, stop_col), daemon=True)
    col.start()

    def background_clients(lb_addr, run_seconds=BASELINE_DURATION+10, rps=10):
        end = time.time() + run_seconds
        s = requests.Session()
        interval = 1.0/max(1, rps)
        while time.time() < end:
            try:
                s.get(f"http://{lb_addr}/", timeout=2)
            except:
                pass
            time.sleep(interval)

    bg = threading.Thread(target=background_clients, args=(f"127.0.0.1:{lb.port}", BASELINE_DURATION+10, 10), daemon=True)
    bg.start()

    time.sleep(BASELINE_DURATION + 2)
    at = threading.Thread(target=attacker_sim, args=(f"127.0.0.1:{lb.port}", ATTACK_DURATION, ATTACKER_RATE_ATTACK), daemon=True)
    at.start()

    total_time = max(60, BASELINE_DURATION + ATTACK_DURATION + 30)
    start = time.time()
    timeline = []
    while time.time() - start < total_time:
        timeline.append((time.time(), len(lb.backends), len(lb.blacklist)))
        time.sleep(1)

    stop_col.set()
    monitor.stop()
    for ev in backend_events.values():
        ev.set()

    with open(os.path.join(OUTPUT_DIR,"timeline.csv"), "w") as f:
        w = csv.writer(f)
        w.writerow(["ts","backends_len","blacklist_len"])
        for r in timeline:
            w.writerow(r)

    with open(os.path.join(OUTPUT_DIR,"monitor_actions.csv"), "w") as f:
        wcsv = csv.writer(f)
        wcsv.writerow(["ts","action","v1","v2","v3","v4"])
        for a in monitor.actions_log:
            wcsv.writerow(list(a) + [""]*(6-len(a)))

    print("[demo] finished. artifacts in ./output")

    try:
        df = pd.read_csv(os.path.join(OUTPUT_DIR,"metrics_log.csv"), names=["ts","reqs","errs"])
        df['t'] = df['ts'] - df['ts'].iloc[0]
        plt.figure(figsize=(10,3))
        plt.plot(df['t'], df['reqs'])
        plt.title("requests per MONITOR_INTERVAL")
        plt.xlabel("seconds")
        plt.grid(True)
        plt.show()

        plt.figure(figsize=(10,3))
        plt.plot(df['t'], df['errs'])
        plt.title("errors per MONITOR_INTERVAL")
        plt.xlabel("seconds")
        plt.grid(True)
        plt.show()
    except Exception as e:
        print("plot error", e)

# clear output files and run
open(os.path.join(OUTPUT_DIR,"metrics_log.csv"), "w").close()
run_demo()
```
